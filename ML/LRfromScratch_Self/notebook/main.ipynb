```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
plt.figure(figsize=(10, 6)) 
```

```python
train=pd.read_csv('../data/train.csv')
test=pd.read_csv('../data/test.csv')
```

```python
test=test.dropna()
train=train.dropna()
```

```python
plt.scatter(x=train['x'],y=train['y'])
```

```python
plt.scatter(x=test['x'],y=test['y'])
```

```python
xtrain=train[['x']].values
xtest=test[['x']].values
ytrain=train[['y']].values
ytest=test[['y']].values
```

```python
xtrain.shape
```

```python
W=0
B=0
m=len(ytrain)
lr=.0001
iter=1000
costs=[]
```

```python
def compute_cost(x,y,W,B):
    pred=x*W+B
    err=pred-y
    cost=np.sum(err**2)/(2 * m)
    return cost
```

```python
for i in range(iter):
    pred=xtrain*W+B
    err=pred-ytrain
    gradientB=(1/m)*np.sum(err)
    gradientW=(1/m)*np.sum(err*xtrain)

    W=W-lr*gradientW
    B=B-lr*gradientB

    cost=compute_cost(xtrain,ytrain,W,B)
    costs.append(cost)
print(costs)
```

```python
testcost=compute_cost(xtest,ytest,W,B)
testcost
```

```python
plt.plot(range(iter), costs)
plt.title("Cost Function Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
#final predictions
final_predictions = W * xtest + B

plt.scatter(xtest, ytest, label='Original Data')
plt.plot(xtest, final_predictions, color='red', label='Regression Line')
plt.title("Final Model Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
